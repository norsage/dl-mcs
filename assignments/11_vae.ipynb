## Variational autoencoders (VAE)

Выберите датасет, на котором хотите обучать вашу генеративную модель:
- CelebA: https://www.kaggle.com/datasets/jessicali9530/celeba-dataset
- PeopleArt: https://www.kaggle.com/datasets/amanagr/people-art-dataset
- Best Artworks of All Time: https://www.kaggle.com/datasets/ikarus777/best-artworks-of-all-time
- Cartoon Faces (Google's Cartoon Set): https://www.kaggle.com/datasets/brendanartley/cartoon-faces-googles-cartoon-set

#### Задание 1 (3 балла). Обучение VAE

Используйте реализацию VAE с последней [практики](../workshops/13_vae.ipynb) для обучения двух вариантов моделей:
1. с параметром `beta > 0.005`
2. с параметром `beta < 0.0001`

Обучите обе модели в течение >50000 шагов, сгенерируйте 64 изображения для каждой, опишите видимые различия

#### Задание 2 (3+2 балла). Интерполяция между изображениями

Сделайте интерполяцию между двумя латентными векторами и сгенерируйте изображения лиц для полученной последовательности. Вам пригодятся функции `torch.lerp` и `torch.linspace`.

Примерная схема:
1. Выберите два изображения из датасета, между которыми хотите сделать интерполяцию
2. Получите их латентные представления $z_0$ и $z_1$ с помощью энкодера
3. Постройте последовательность 10 интерполированных значений $z_t$, $t \in [0, 1]$
4. Для всех $z_t$ сгенерируйте изображения с помощью декодера
5. Отрисуйте все полученные изображения в один ряд

Бонус (+2 балла): реализуйте сферическую линейную интерполяцию (SLERP), сравните результаты с простой линейной.

#### Задание 3 (12 баллов). VQ-VAE

Реализуйте и обучите VQ-VAE (https://paperswithcode.com/method/vq-vae)

Допускается использовать код, который найдёте на github (есть, например, такая реализация: https://github.com/MishaLaskin/vqvae)

Но можете попробовать свои силы и сделать всё самостоятельно по тексту статьи или материалам лекций — идея и реализация довольно простые, а вы получите бесценный опыт

Оберните реализацию в lightning и запустите на вашем датасете на >100000 шагов, выведите 64 примера сгенерированных изображений
